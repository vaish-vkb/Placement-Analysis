## 1. Which factor influenced the candidate in getting placed?

```python
df['status'].value_counts()
```

```python
for col in df.select_dtypes(include='object'):
    sns.countplot(x=col, hue  ='status', data=df)
    plt.show()
```

```python
for i in df.select_dtypes(include='object').columns:
    ((pd.crosstab(df[i],df['status'],normalize = 'index')).transpose()).plot(kind='pie',subplots=True,figsize =(15,8),title = df[i].name)
```

```python
for i in df.select_dtypes(include='object').columns:
    print(i,'=',df[i].unique())
```

```python
df['hsc_p'].describe()
```

```python
for i in df.columns:
    if '_p' in i:
        sns.barplot(x=df['status'],y=df[i])
        plt.title(i+' vs Status')
        plt.show()
```
